```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import joblib
```

```python
df2 = pd.read_csv('/content/dataset-prediksi-sem2.csv')
df3 = pd.read_csv('/content/dataset-prediksi-sem3.csv')
df4 = pd.read_csv('/content/dataset-prediksi-sem4.csv')
df5 = pd.read_csv('/content/dataset-prediksi-sem5.csv')
df6 = pd.read_csv('/content/dataset-prediksi-sem6.csv')
df7 = pd.read_csv('/content/dataset-prediksi-sem7.csv')
df8 = pd.read_csv('/content/dataset-prediksi-sem8.csv')
```

```python
X2 = df2['ips-now']
X3 = df3['ips-now']
X4 = df4['ips-now']
X5 = df5['ips-now']
X6 = df6['ips-now']
X7 = df7['ips-now']
X8 = df8['ips-now']
```

```python
y2 = df2['ips-next']
y3 = df3['ips-next']
y4 = df4['ips-next']
y5 = df5['ips-next']
y6 = df6['ips-next']
y7 = df7['ips-next']
y8 = df8['ips-next']
```

```python
# Dataset prediksi sem2
X2_train, X2_temp, y2_train, y2_temp = train_test_split(X2, y2, test_size=0.3, random_state=42)
X2_val, X2_test, y2_val, y2_test = train_test_split(X2_temp, y2_temp, test_size=0.5, random_state=42)
X2_train_reshaped = X2_train.values.reshape(-1, 1)
X2_val_reshaped = X2_val.values.reshape(-1, 1)
X2_test_reshaped = X2_test.values.reshape(-1, 1)
```

```python
# Dataset prediksi sem3
X3_train, X3_temp, y3_train, y3_temp = train_test_split(X3, y3, test_size=0.3, random_state=42)
X3_val, X3_test, y3_val, y3_test = train_test_split(X3_temp, y3_temp, test_size=0.5, random_state=42)
X3_train_reshaped = X3_train.values.reshape(-1, 1)
X3_val_reshaped = X3_val.values.reshape(-1, 1)
X3_test_reshaped = X3_test.values.reshape(-1, 1)
```

```python
# Dataset prediksi sem2
X4_train, X4_temp, y4_train, y4_temp = train_test_split(X4, y4, test_size=0.3, random_state=42)
X4_val, X4_test, y4_val, y4_test = train_test_split(X4_temp, y4_temp, test_size=0.5, random_state=42)
X4_train_reshaped = X4_train.values.reshape(-1, 1)
X4_val_reshaped = X4_val.values.reshape(-1, 1)
X4_test_reshaped = X4_test.values.reshape(-1, 1)
```

```python
# Dataset prediksi sem2
X5_train, X5_temp, y5_train, y5_temp = train_test_split(X5, y5, test_size=0.3, random_state=42)
X5_val, X5_test, y5_val, y5_test = train_test_split(X5_temp, y5_temp, test_size=0.5, random_state=42)
X5_train_reshaped = X5_train.values.reshape(-1, 1)
X5_val_reshaped = X5_val.values.reshape(-1, 1)
X5_test_reshaped = X5_test.values.reshape(-1, 1)
```

```python
# Dataset prediksi sem2
X6_train, X6_temp, y6_train, y6_temp = train_test_split(X6, y6, test_size=0.3, random_state=42)
X6_val, X6_test, y6_val, y6_test = train_test_split(X6_temp, y6_temp, test_size=0.5, random_state=42)
X6_train_reshaped = X6_train.values.reshape(-1, 1)
X6_val_reshaped = X6_val.values.reshape(-1, 1)
X6_test_reshaped = X6_test.values.reshape(-1, 1)
```

```python
# Dataset prediksi sem2
X7_train, X7_temp, y7_train, y7_temp = train_test_split(X7, y7, test_size=0.3, random_state=42)
X7_val, X7_test, y7_val, y7_test = train_test_split(X7_temp, y7_temp, test_size=0.5, random_state=42)
X7_train_reshaped = X7_train.values.reshape(-1, 1)
X7_val_reshaped = X7_val.values.reshape(-1, 1)
X7_test_reshaped = X7_test.values.reshape(-1, 1)
```

```python
# Dataset prediksi sem2
X8_train, X8_temp, y8_train, y8_temp = train_test_split(X8, y8, test_size=0.3, random_state=42)
X8_val, X8_test, y8_val, y8_test = train_test_split(X8_temp, y8_temp, test_size=0.5, random_state=42)
X8_train_reshaped = X8_train.values.reshape(-1, 1)
X8_val_reshaped = X8_val.values.reshape(-1, 1)
X8_test_reshaped = X8_test.values.reshape(-1, 1)
```

```python
model = MLPRegressor(hidden_layer_sizes=(8, 8, 8, 8, 8, 8, 8), max_iter=100, random_state=42, solver='lbfgs', early_stopping=True, n_iter_no_change=10)
```

```python
# Latih, validasi, dan uji model sem2
training_sem2 = model.fit(X2_train_reshaped, y2_train)
validating_sem2 = model.predict(X2_val_reshaped)
testing_sem2 = model.predict(X2_test_reshaped)
```

```python
#evaluasi pada data validasi
mse_val_sem2 = mean_squared_error(y2_val, validating_sem2)
r2_val_sem2 = r2_score(y2_val, validating_sem2)

#evaluasi pada data uji
mse_test_sem2 = mean_squared_error(y2_test, testing_sem2)
r2_test_sem2 = r2_score(y2_test, testing_sem2)
```

```python
# Menampilkan hasil evaluasi
print(f"MSE (Validation): {mse_val_sem2}")
print(f"R-squared (Validation): {r2_val_sem2}")
print(f"MSE (Test): {mse_test_sem2}")
print(f"R-squared (Test): {r2_test_sem2}")
```

```python
# Plot hasil prediksi pada data validasi dan data uji
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.scatter(y2_val, validating_sem2)
plt.title(f"MLP Regressor - Semester 2 - Validation Set")
plt.xlabel("True Values")
plt.ylabel("Predictions")
```

```python
# Plot hasil prediksi pada data validasi dan data uji
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.scatter(y2_test, testing_sem2)
plt.title(f"MLP Regressor - Semester 2 - Testing Set")
plt.xlabel("True Values")
plt.ylabel("Predictions")
```

```python
# Grafik Residual
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
residuals = y2_test - testing_sem2
plt.scatter(testing_sem2, residuals)
plt.title("Residual Plot Semester 2")
plt.xlabel("Predictions")
plt.ylabel("Residuals")
```

```python
# Grafik Normalitas Residual
plt.subplot(2, 2, 2)
sns.histplot(residuals, kde=True)
plt.title("Residual Normality Plot Semester 2")
plt.xlabel("Residuals")
```
